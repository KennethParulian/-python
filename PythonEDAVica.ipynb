#Kelas Python EDA


####Import Library

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

#### Connect Datasource

```python
link = 'https://raw.githubusercontent.com/KennethParulian/PythonEDAEnigmacamp/main/MOCK_DATA%20(4).csv'
df = pd.read_csv(link)
```

####Data Validation

```python
df.head(20)
```

```python
df['CreditRemaining'] = np.random.randint(1000, 15000, df.shape[0])
df.head(10)
```

####Information of Dataset

```python
df.info()
```

####Duplicated Data Handling

```python
df.duplicated(subset=None, keep='first')
```

####Drop Unnecesary Data

```python
df.drop("email", axis=1, inplace=True)
```

```python
df.head (10)
```

####Check Null Data

```python
print(df.isnull().sum())
```

####Clean Null Data

```python
df = df.dropna()    
df.count()
```

```python
df["CreditCardLimit"] = df["CreditCardLimit"].replace('[\$\,\.]',"",regex=True).astype(int)
df
```

```python
df = df.rename(columns={"first_name": "FirstName", "last_name": "LastName", "gender": "Gender"})
df.head()
```

```python
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(x="CreditCardLimit", y = df.CreditCardLimit.index , data=df);

```

```python
plt.barh(df['id'],df['CreditCardLimit'],color=['blue'])
plt.grid(linestyle="--", linewidth=2, axis='x', alpha = 0.5)
plt.ylabel("Credit Card Limit")
plt.xlabel("CreditCardNumber")
plt.title("Credit Card Limit Berdasarkan Credit Card Number")

fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(40)
```
